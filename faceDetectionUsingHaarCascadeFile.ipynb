```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face_cas = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread(r"C:\Users\sharo\Desktop\OpenCV\group.jpg")
plt.imshow(img)
```

```python
image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#converting to grey (matplotlib use cheyth grey convertion cheyth view cheythal yellow color aayirikkum)

img_grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("img",img_grey)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
faces = face_cas.detectMultiScale(img_grey,scaleFactor=1.1,minNeighbors=4,minSize=(30,30))
```

```python
faces
```

```python
#rectangle 

for (x,y,w,h) in faces:
    cv2.rectangle(img_grey,(x,y),(x+w,y+h),color=(0,255,0),thickness=3)
```

```python
cv2.imshow("img",img_grey)
cv2.waitKey(5000)
cv2.destroyAllWindows()
    
```

```python
while True:
    success,frames=video.read()
    
    grey = cv2.cvtColor(frames,cv2.COLOR_BGR2GRAY)
    
    faces =  face_cas.detectMultiScale(grey,scaleFactor=1.1,minNeighbors=4,minSize=(30,30))
    
    # print(faces)
    
    for (x,y,w,h) in faces:
        cv2.rectangle(frames,(x,y),(x+w,y+h),(255,0,0),3)
    
    cv2.imshow("video",frames)
    if cv2.waitKey(1) & 0xFF ==27:
        break

video.release()
cv2.destroyAllWindows()
```
