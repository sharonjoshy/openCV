```python
import cv2 as CV

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
eye_img = CV.imread("ronaldo.jpg")
```

```python
eye_img
```

```python
eye_cas = CV.CascadeClassifier("haarcascade_eye.xml")
```

```python
eyes = eye_cas.detectMultiScale(eye_img,1.1,4,minSize=(30,30))
```

```python
eyes
```

```python
len(eyes)
```

```python
for (x,y,w,h) in eyes:
    CV.rectangle(eye_img,(x,y),(x+w,y+h),(50,40,20),2)
```

```python
plt.imshow(eye_img)
```

# detecting eyes using webcam


```python
cap = CV.VideoCapture(0)
```

```python

while True:
    ret,frame = cap.read()   #convertig ito frame
    grey = CV.cvtColor(frame,CV.COLOR_BGR2GRAY)   # frame ilek convert cheythath grey aakunu
    faces = eye_cas.detectMultiScale(grey,1.1,4,minSize=(30,200))
    for (x,y,w,h) in faces:
        CV.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)

    #image = CV.resize(frame, (200,200))  #size kurakkan
    CV.imshow("video",frame)  # for playing video
    if CV.waitKey(1)& 0xFF ==27:
        break
cap.release()
CV.destroyAllWindows()
```
